```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x=[1.25,2.5,3.75,5.,1.25]
alum=[2077.,1670.,1362.67,1038.,2038.66]
co=[1219.,632.,317.67,173.,1284.67]
lead=[797.66,315.33,149.,75.,792.67]

mu_web=[0.35515,0.62030,1.19189]

mu_alum=[]
mu_co=[]
mu_lead=[]

Ni=2555.33
# N


#calculate mu by Mu=-(log(N/N_0)/x)

print("alum mu calculation")
for i in range(0,len(alum)):
    mu=-np.log(alum[i]/Ni)/(x[i])
    print("i:"+str(i) + " x: " + str(x[i]) + " mu: "+str(mu))
    mu_alum.append(mu)
    
print("co mu calculation")
for i in range(0,len(co)):
    mu=-np.log(co[i]/Ni)/(x[i])
    print("i:"+str(i) + " x: " + str(x[i]) + " mu: "+str(mu))
    mu_co.append(mu)
    
print("lead mu calculation")
for i in range(0,len(lead)):
    mu=-np.log(lead[i]/Ni)/(x[i])
    print("i:"+str(i) + " x: " + str(x[i]) + " mu: "+str(mu))
    mu_lead.append(mu)

```

```python
plt.figure(figsize=(10, 10))
ax = plt.axes()
ax.scatter(x,mu_alum)
ax.plot([0,5.], [mu_web[0],mu_web[0]])

plt.figure(figsize=(10, 10))
ax = plt.axes()
ax.scatter(x,mu_co)
ax.plot([0,5.], [mu_web[1],mu_web[1]])

plt.figure(figsize=(10, 10))
ax = plt.axes()
ax.scatter(x,mu_lead)
ax.plot([0,5.], [mu_web[2],mu_web[2]])
```

$\chi^2$ Minimization

```python

N0=2555.33
al_ar=[2077,1670,1362.67,1038]
co_ar=[1219,632,317.67,173]
pb_ar=[797.66,315.33,149,75]
x_arr=[1.25,2.5,3.75,5.]
al_arry=np.array(al_ar)/N0
co_arry=np.array(co_ar)/N0
pb_arry=np.array(pb_ar)/N0

al_uncer=[60.11,12.12,20.79,18.33]
co_uncer=[12.12,10.82,11.22,2.89]
pb_uncer=[16.86,20.84,7.55,4.51]


def my_fun(inter_mu,name):
    
    inter=inter_mu[0]
    mu=inter_mu[1]
    
    N0=2555.33
    al_ar=[2077,1670,1362.67,1038]
    co_ar=[1219,632,317.67,173]
    pb_ar=[797.66,315.33,149,75]
    x_arr=[1.25,2.5,3.75,5.]
    al_arry=np.array(al_ar)/N0
    co_arry=np.array(co_ar)/N0
    pb_arry=np.array(pb_ar)/N0

    al_uncer=[60.11,12.12,20.79,18.33]
    co_uncer=[12.12,10.82,11.22,2.89]
    pb_uncer=[16.86,20.84,7.55,4.51]


    chi_al=0
    chi_cu=0
    chi_pb=0
    
    x_arr=np.array(x_arr)
    
    exps=-1*np.array(x_arr)*mu+inter
    model_ar=np.exp(exps)*N0
    
    for i in range(0,4):
        chi_al+=((model_ar[i]-al_ar[i])**2)/((al_uncer[i])**2)
        chi_cu+=((model_ar[i]-co_ar[i])**2)/((co_uncer[i])**2)
        chi_pb+=((model_ar[i]-pb_ar[i])**2)/((pb_uncer[i])**2)
    

    if name=="al":
        #     print("Aluminmum")
        #     print("model_ar")
        #     print(model_ar)
        #     plt.scatter(x_arr,al_ar,color="green")
        #     plt.scatter(x_arr,model_ar)
        result=chi_al
        
    if name=="cu":
        result= chi_cu
    if name == "pb":
        result = chi_pb

    return result
    


z=my_fun(inter_mu=[0.030507091908313133,0.18274111],name="al")

print(z)
```

```python
import scipy

x0=(0.030507091908313133,0.18274111)

print(scipy.optimize.minimize(my_fun,x0,method="SLSQP",args="al"))
print(" ")
print("Al Chi^2 minimization: chi^2=1.1003383527890525, intercept=0.04061259, mu=0.18592093")
```

```python
#CU

# Co
# coefficient of determination: 0.9994298103749627
# intercept: -0.09235162350447212
# slope: [-0.52362879]

x0=(-0.09235162350447212,0.52362879)
print(scipy.optimize.minimize(my_fun,x0,method="SLSQP",args="cu"))

print()

print("Cu Chi^2 minimization: chi^2=0.3696422948282022, intercept=-0.08923204, mu=0.5221143")
```

```python
# Pb
# coefficient of determination: 0.9950631284701559
# intercept: -0.44618841333140113
# slope: [-0.62738051]

x0=(-0.44618841333140113, 0.82738051)
print(scipy.optimize.minimize(my_fun,x0,method="SLSQP",args="pb"))

print()

print("Pb Chi^2 minimization: chi^2=2.7533458484907465, intercept=-0.08923204, mu=0.5221143")
```
